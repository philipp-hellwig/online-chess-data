```python
import pandas as pd
from datetime import date
import numpy as np
```

```python
df = pd.read_csv("../data/blitz_n50_Aug24.csv")
# convert date strings to date time:
df["createdAt"] = df["createdAt"].apply(pd.Timestamp, unit="D")
df["lastMoveAt"] = df["lastMoveAt"].apply(pd.Timestamp, unit="D")

#recalculate outcomes
df["white"] = df["white"].apply(str)
df["username"] = df["username"].apply(str)
df["black"] = df["black"].apply(str)
df["white"] = df["white"].apply(str.lower)
df["username"] = df["username"].apply(str.lower)
df["black"] = df["black"].apply(str.lower)
# add outcome of the game (win, draw, loss) from the perspective of the username:
conditions = [
    df["winner"].isna(),
    (df["winner"]=="white") & (df["white"]==df["username"]),
    (df["winner"]=="black") & (df["black"]==df["username"]),
]
choices = [
    "draw",
    "win",
    "win"
]
df["outcome"] = np.select(conditions, choices, default="loss")

# add date by day:
df["createdAt"].apply(lambda x: x.strftime('%Y-%m-%d'))
df["date_created"] = df["createdAt"].apply(lambda x: x.strftime('%Y-%m-%d'))
```

```python
performance_by_day = df.groupby(["username","date_created"])["outcome"].value_counts()
```
